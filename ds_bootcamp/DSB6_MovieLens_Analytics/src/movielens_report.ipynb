### 3. Голос пользователей: что им нравится?

```python
%timeit
ratings = ml.Ratings(data_path + 'ratings.csv')
ratings_movies = ml.Ratings.Movies(ratings)
ratings_users = ml.Ratings.Users(ratings)

ratings_movies.dist_by_rating()
```

#### 3.1. Когда пользователи были наиболее активны?

```python
%timeit
ratings_movies.dist_by_year()
```

#### 3.2. Какие оценки ставят чаще всего?

```python
%timeit
ratings_movies.dist_by_rating()
```

#### 3.3. Какие фильмы обсуждают больше всего?

```python
%timeit
ratings_movies.top_by_num_of_ratings(5)
```

#### 3.4. Какие фильмы получили самые высокие оценки? 
(по оценке среднего: чувствительно к выбросам, аномалиям)

```python
%timeit
ratings_movies.top_by_ratings(5, metric='average')
```

#### 3.5. А если посмотреть на медиану?
Медиана менее чувствительна к выбросам.

```python
%timeit
ratings_movies.top_by_ratings(5, metric='median')
```

#### 3.6. Какие фильмы вызывают самые жаркие споры?
Высокая дисперсия = больше разных мнений.

```python
%timeit
ratings_movies.top_controversial(5)
```

#### 3.7. А если учитывать только популярные фильмы? (бонус)
Фильмы с достаточным числом оценок.

```python
%timeit
ratings_movies.top_by_num_of_ratings_threshold(4)
```

Самые высоко оцененные фильмы при фильтрации

```python
%timeit
ratings_movies.top_by_ratings_filtered(5, metric='average', min_ratings=4)
```
